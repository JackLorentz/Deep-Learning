```python
import numpy as np
import tensorflow as tf 
import matplotlib.pyplot as plt
%matplotlib inline

import cv2
import random
import keras
import time
import os

from scipy.io import loadmat
from tf_pose import common
from tf_pose.estimator import TfPoseEstimator
from tf_pose.networks import get_graph_path, model_wh
from sklearn import metrics
from random import randint
from os import listdir
from os.path import isfile, isdir, join
```

```python
w, h = model_wh('432x368')
e = TfPoseEstimator(get_graph_path('mobilenet_thin'), target_size=(432, 368))
```

```python
x_data = []
y_data = []
```

```python
path = './data'
classList = ['lying/self_lying_data', 'fall_floor', 'stand', 'walk', 'sit']
for i in range(len(classList)):
    classPath = join(path, classList[i])
    files = listdir(classPath)
    for f in files:
        fullpath = join(classPath, f)
        frame_cnt = 0
        cap = cv2.VideoCapture(fullpath)
        success, img = cap.read()
        #最多就取32幀
        action = np.zeros([32, 36])
        while success and frame_cnt < 32:
            img = cv2.resize(img, (432, 368))
            width, height, channel = img.shape
            human = e.inference(img, resize_to_default=(w>0 and h>0), upsample_size=4.0)
            #若有偵測到骨架
            if human:
                for part_idx in human[0].body_parts:
                    action[frame_cnt, part_idx*2] = human[0].body_parts[part_idx].x*width + 0.5
                    action[frame_cnt, part_idx*2+1] = human[0].body_parts[part_idx].y*height + 0.5
            print("The number", frame_cnt, "of frame in " + f)

            cap.set(cv2.CAP_PROP_POS_MSEC, (frame_cnt*85))#可擷取到32個frames
            success, img = cap.read()
            print('Read a new frame: ', success)
            frame_cnt += 1
        #不足32幀,就補到32幀,使每一個動作都有32幀
        last = frame_cnt
        while frame_cnt < 31:
            action[frame_cnt, :] = action[last, :]
            frame_cnt += 1

        x_data.append(action)
        y_data.append(i)
        cap.release()
```

```python
classList = ['lying/self_lying_data']
for i in range(len(classList)):
    classPath = join(path, classList[i])
    files = listdir(classPath)
    for f in files:
        if f.find('.mp4') > -1 or f.find('.avi') > -1:
            fullpath = join(classPath, f)
            frame_cnt = 0
            cap = cv2.VideoCapture(fullpath)
            success, img = cap.read()
            #最多就取32幀
            action = np.zeros([32, 36])
            while success and frame_cnt < 32:
                img = cv2.resize(img, (432, 368))
                img = cv2.flip(img, 1)
                human = e.inference(img, resize_to_default=(w>0 and h>0), upsample_size=4.0)
                #若有偵測到骨架
                if human:
                    width, height, channel = img.shape
                    for part_idx in human[0].body_parts:
                        action[frame_cnt, part_idx*2] = human[0].body_parts[part_idx].x*width + 0.5
                        action[frame_cnt, part_idx*2+1] = human[0].body_parts[part_idx].y*height + 0.5
                print("The number", frame_cnt, "of frame in " + f)

                cap.set(cv2.CAP_PROP_POS_MSEC, (frame_cnt*85))#可擷取到32個frames
                success, img = cap.read()
                print('Read a new frame: ', success)
                frame_cnt += 1
            #不足32幀,就補到32幀,使每一個動作都有32幀
            last = frame_cnt
            while frame_cnt < 31:
                action[frame_cnt, :] = action[last, :]
                frame_cnt += 1

            lying.append(action)
            cap.release()
```
